```python
import numpy as np
import matplotlib.pyplot as plt
import csv


t = t = np.linspace(0,1,10)
y = (t**2 + 2*t + 4) / 2    #original solution
# Reading the CSV file
t1 = []
y1 = []
with open('rk_output1.csv', 'r') as file:
    data = csv.DictReader(file)
    for row in data:
        t1.append(float(row['t'].strip()))
        y1.append(float(row[' y'].strip()))
t2 = []
y2 = []
with open('rk_output2.csv', 'r') as file:
    data = csv.DictReader(file)
    for row in data:
        t2.append(float(row['t'].strip()))
        y2.append(float(row[' y'].strip()))

# Plotting the data
plt.figure(figsize=(8, 8))
plt.plot(t, y, marker='o', linestyle='--', color='g', label='original solution')
plt.plot(t1, y1, marker='*', linestyle='-', color='b', label='for h=0.1')
plt.plot(t2, y2, marker='o', linestyle='-', color='r', label='for h=0.05')

plt.title('Plot of y vs t for RK 4th order')
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```

**Conclusion:**
In RK 4th order, all the three coincides.
